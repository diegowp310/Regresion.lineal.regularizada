```python
# Your code here

import pandas as pd

df = pd.read_csv("https://breathecode.herokuapp.com/asset/internal-link?id=733&path=demographic_health_data.csv")
```

```python
df
```

```python
pd.set_option('display.max_columns', None)

print(df.head())
```

```python
df['COUNTY_NAME'] = pd.factorize(df['COUNTY_NAME'])[0]
df['STATE_NAME'] = pd.factorize(df['STATE_NAME'])[0]

df.info()
```

```python
X = df.drop('Obesity_number', axis=1)
Y = df['Obesity_number']
```

```python
correlaciones = X.corrwith(Y)
print(correlaciones)
```

```python
altas_correlaciones = correlaciones[correlaciones>0.95]
print(altas_correlaciones)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_esc = scaler.fit_transform(X_train, Y_train)
X_test_esc = scaler.transform(X_test)

X_train_esc = pd.DataFrame(X_train_esc, index = X_train.index, columns=X_train.columns)
X_test_esc = pd.DataFrame(X_test_esc, index = X_test.index, columns=X_test.columns)
```

```python
from sklearn.linear_model import LinearRegression, Lasso

modelRL= LinearRegression()
modelLasso=Lasso()

modelRL.fit(X_train_esc, Y_train)
modelLasso.fit(X_train_esc, Y_train)
```

```python
Y_pred_RL = modelRL.predict(X_test_esc)
Y_pred_Lasso = modelLasso.predict(X_test_esc)

from sklearn.metrics import r2_score, root_mean_squared_error

print(f'Para la Regresion lineal se obtuvo un RMSE de: {root_mean_squared_error(Y_test, Y_pred_RL)} y un R2 de: {r2_score(Y_test, Y_pred_RL)}')
print(f'Para la Regresion lineal se obtuvo un RMSE de: {root_mean_squared_error(Y_test, Y_pred_Lasso)} y un R2 de: {r2_score(Y_test, Y_pred_Lasso)}')
```

```python
import numpy as np
espacio = np.linspace(0,20,40)
print(espacio)
```

```python
R2=[]
variables_activas =[]
for variable in espacio:
    model_Lasso_variable = Lasso(alpha=variable, max_iter=100000)
    model_Lasso_variable.fit(X_train_esc, Y_train)
    Y_pred = model_Lasso_variable.predict(X_test_esc)
    R2.append(r2_score(Y_test, Y_pred))
    cantidad_variables_distintas_cero = np.sum(model_Lasso_variable.coef_ !=0)
    variables_activas.append(cantidad_variables_distintas_cero)
print(R2)
```

```python
import matplotlib.pyplot as plt 

plt.plot(espacio,R2)
plt.show 
```

```python
plt.plot(espacio, variables_activas)
plt.show
```
